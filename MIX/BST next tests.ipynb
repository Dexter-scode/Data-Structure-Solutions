```python
class Node:
    def __init__(self,value=None):
        self.value=value
        self.left=None
        self.right=None

def insert(A,data):
    q=[]
    q.append(A)
    while(len(q)):
        temp=q.pop(0)
        if(temp and temp.value>data):
            if(temp.left==None):
                temp.left=Node(data)
            else:
                q.append(temp.left)
        elif(temp and temp.value<data):
            if(temp.right==None):
                temp.right=Node(data)
            else:
                q.append(temp.right)
        else:
            print("SORRY NUMBER ALREADY PRESENT")

def morrisorder(root):
    cur=root
    while(cur):
        if(cur.left == None):
            print(cur.value,end=" ")
            cur=cur.right
        else:
            pre=cur.left
            while(pre.right and pre.right != cur):
                pre=pre.right
            if(pre.right==None):
                pre.right=cur
                cur=cur.left
            else:
                pre.right=None
                print(cur.value,end=" ")
                cur=cur.right
def inorder(A):
    cur=A
    stack=[]
    ele=[]
    while(True):
        if(cur):
            stack.append(cur)
            cur=cur.left
        elif(stack):
            cur=stack.pop()
            ele.append(cur.value)
            cur=cur.right
        else:
            break
    return(ele)
def height(A):
    if(A==None):
        return(0)
    else:
        l=height(A.left)
        r=height(A.right)
        return(max(l,r)+1)
    
def levelorderusingqueue(A):
    ele=[]
    q=[]
    q.append(A)
    while(len(q)):
        temp=q.pop(0)
        if(temp):
            ele.append(temp.value)
            if(temp.left):
                q.append(temp.left)
            if(temp.right):
                q.append(temp.right)
    return(ele)
def pathuntil(A,data,arr):
    if(not A):
        return(False)
    arr.append(A.value)
    if(A.value==data):
        return(True)
    if(pathuntil(A.left,data,arr) or pathuntil(A.right,data,arr)):
        return(True)
    arr.pop()
    return(False)
    
def path(A,data):
    arr=[]
    pathuntil(A,data,arr)
    return(arr)

tree=None
A=[4,2,6,1,3,5,7]
for i in range(len(A)):
    if(tree is None):
        tree=Node(A[i])
    else:
        insert(tree,A[i])
```
