```python
class Node:
    def __init__(self,value=None):
        self.value=value
        self.left=None
        self.right=None
class BST:
    def __init__(self):
        self.root=None
    
    def insert(self,data):
        newnode=Node(data)
        if(self.root==None):
            self.root=newnode
        else:
            self._insert(data,self.root)
    
    def _insert(self,value,cur_node):
        if(value<cur_node.value):
            if(cur_node.left==None):
                cur_node.left=Node(value)
            else:
                self._insert(value,cur_node.left)
        elif(value>cur_node.value):
            if(cur_node.right==None):
                cur_node.right=Node(value)
            else:
                self._insert(value,cur_node.right)
        else:
            print("Value Alredy in tree")
    def height(self):
        if(self.root!=None):
            return(self._height(self.root,0))
        else:
            return(0)
        
    def _height(self,cur_node,cur_height):
        if(cur_node==None):
            return(cur_height)
        left_height=self._height(cur_node.left,cur_height+1)
        right_height=self._height(cur_node.right,cur_height+1)
        return(max(left_height,right_height))
    
    def print(self):
        ele=[]
        if(self.root==None):
            return([])
        else:
            self._print(self.root,ele)
            return(ele)
    def _print(self,cur_node,ele):
        if(cur_node!=None):
            print(cur_node.value)
            self._print(cur_node.left,ele)
            ele.append(cur_node.value)
            self._print(cur_node.right,ele)
            
    def perfect(self):
        A=self.print()
        self.root=self._perfect(A,0,len(A)-1)
    def _perfect(self,A,start,end):
        if(start>end):
            return()
        if(start==end):
            return(Node(A[start]))
        mid=start+(end-start)//2
        cur=Node(A[mid])
        cur.left=self._perfect(A,start,mid-1)
        cur.right=self._perfect(A,mid+1,end)
        return(cur)
        
tree=BST()
tree.insert(3)
tree.insert(2)
tree.insert(1)
tree.insert(7)
tree.insert(6)
tree.insert(5)
tree.insert(4)
print(tree.print())
tree.perfect()
print(tree.print())
print(tree.height())
tree.perfect()
print(tree.print())
```
