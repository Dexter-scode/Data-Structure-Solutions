```python
class Node:
    def __init__(self,value=None):
        self.value=value
        self.left=None
        self.right=None

        
def insert(A,data):
    q=[]
    q.append(A)
    while(len(q)):
        temp=q.pop(0)
        if(not temp.left):
            temp.left=Node(data)
            break
        else:
            q.append(temp.left)
        if(not temp.right):
            temp.right=Node(data)
            break
        else:
            q.append(temp.right)

def height(A):
    if(A==None):
        return(0)
    else:
        l=height(A.left)
        r=height(A.left)
        return(max(l,r)+1)
    
def reverselevelorder(A,l,arr):
    if(A == None):
        return(None)
    elif(l==1):
        arr.append(A.value)
    elif(l>1):
        reverselevelorder(A.left,l-1,arr)
        reverselevelorder(A.right,l-1,arr)

def printreverselevelorder(A):
    new=[]
    h=height(A)
    for i in range(1,h+1):
        arr=[]
        reverselevelorder(A,i,arr)
        arr=arr[::-1]
        new+=arr
    return(new)

def preorder(A):
    if(A):
        preorder(A.left)
        preorder(A.right)
        print(A.value,end=" ")
        
def morrispreorder(A):
    cur=A
    while(cur):
        if(cur.left==None):
            print(cur.value,end=" ")
            cur=cur.right
        else:
            pre=cur.left
            while(pre.right!=None and pre.right!=cur):
                pre=pre.right
            
            if(pre.right==cur):
                pre.right=None
                cur=cur.right
            
            else:
                print(cur.value,end=" ")
                pre.right=cur
                cur=cur.left
                
def postorderusingtwoqueue(A):
    q1=[]
    q2=[]
    q1.append(A)
    while(q1):
        temp=q1.pop()
        q2.append(temp)
        if(temp.left):
            q1.append(temp.left)
        if(temp.right):
            q1.append(temp.right)
    while(q2):
        temp=q2.pop()
        print(temp.value,end=" ")
        
def peek(stack):
    if(stack):
        return(stack[-1])
    else:
        return(None)
    
def postorderusingonequeue(root):
    ans=[]
    if(root==None):
        return()
    stack=[]
    while(True):
        while(root):
            if(root.right):
                stack.append(root.right)
            stack.append(root)
            root=root.left
        root=stack.pop()
        if(root.right != None and peek(stack)==root.right):
            stack.pop()
            stack.append(root)
            root=root.right
        else:
            ans.append(root.value)
            root=None
        if(len(stack)<=0):
            return(ans)
            break
def postorderset(A):
    temp=A
    visited=set()
    while(temp and temp not in visited):
        if(temp.left and temp.left not in visited):
            temp=temp.left
        elif(temp.right and temp.right not in visited):
            temp=temp.right
        else:
            print(temp.value,end=" ")
            visited.add(temp)
            temp=A
            
def printtreeleaves(A):
    if(A==None):
        return()
    if(A.left==None and A.right == None):
        print(A.value)
    printtreeleaves(A.left)
    printtreeleaves(A.right)
    
def preorder(A):
    if(A):
        print(A.value,end=" ")
        preorder(A.left)
        preorder(A.right)
        
def inorder(A):
    if(A):
        inorder(A.left)
        print(A.value,end=" ")
        inorder(A.right)

tree=None
A=[1,2,3,4,5,6,7]
for i in range(len(A)):
    if(tree is None):
        tree=Node(A[i])
    else:
        insert(tree,A[i])
        
inorder(tree)
print()
postorderset(tree)
        
new_A=printreverselevelorder(tree)
newtree=None
for i in range(len(new_A)):
    if(newtree is None):
        newtree=Node(new_A[i])
    else:
        insert(newtree,new_A[i])
        
```
