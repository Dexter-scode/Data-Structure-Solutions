```python
class Node:
    def __init__(self,value=None):
        self.value=value
        self.left=None
        self.right=None

def preorder(root):
    if(root!=None):
        print(root.value,end=" ")
        preorder(root.left)
        preorder(root.right)

def gettree(root,start,end):
    tree=[]
    if(start>end):
        tree.append(None)
        return tree
    for i in range(start,end+1):
        ltrees=gettree(root,start,i-1)
        rtrees=gettree(root,i+1,end)
        for j in ltrees:
            for k in rtrees:
                node=Node(root[i])
                node.left=j
                node.right=k
                tree.append(node)
    return(tree)

A=[4,2,5,1,6,7,3]
n=len(A)
trees=gettree(A,0,n-1)
print(len(trees))
for i in trees:
    preorder(i)
    print()
```
